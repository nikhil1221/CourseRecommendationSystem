```python
import json
import pandas as pd

# Opening JSON file 
f = open('students.json',encoding='utf8') 
  
# returns JSON object as  
# a dictionary 
data = json.load(f) 
f.close()
students=pd.DataFrame(data)
students.drop_duplicates(['sid'])
students.head()
```

```python
del students['targets']
del students['target_super_groups']
# Opening JSON file 
f = open('test_submission.json',encoding='utf8') 
  
# returns JSON object as  
# a dictionary 
data = json.load(f) 
f.close()
test_submissions=pd.DataFrame(data)
test_submissions.isnull().sum()

testresults=pd.merge(test_submissions,students,left_on="studentId",right_on="sid")
testresults.head()
```

```python
f = open('tests_details.json',encoding='utf8') 
  
# returns JSON object as  
# a dictionary 
data = json.load(f) 
f.close()
test_details=pd.DataFrame(data)
test_details.head()

testfirstmatrix=pd.merge(testresults,test_details,left_on="testId",right_on="test_id")
del testfirstmatrix['sid']
del testfirstmatrix['test_id']
# testfirstmatrix.isnull().sum()
testfirstmatrix.head()
```

```python
f = open('course_enrollments.json',encoding='utf8') 
  
# returns JSON object as  
# a dictionary 
data = json.load(f) 
f.close()
course_enrollments=pd.DataFrame(data)

#course enrollment destructre

course_enrollments=course_enrollments.enrolled_courses.apply(pd.Series).merge(course_enrollments,left_index=True,right_index=True).drop(['enrolled_courses'],axis=1).melt(id_vars=['studentIds'],value_name='enrolled_courses').drop('variable',axis=1).dropna()
course_enrollments.head()
```

```python
#remove null values from the course
# course_enrollments.isnull().sum()
#course_enrollments[course_enrollments['enrolled_courses'].isnull()]
#course_enrollments.describe()


testmatrix=pd.merge(course_enrollments,students,left_on="studentIds",right_on="sid",how="left")
testmatrix.head()
```
